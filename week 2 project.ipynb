```python
housing_median_age_series = df['housing_median_age']
median_income_series = df['median_income']
median_house_value_series = df['median_house_value']
print("\nHousing Median Age Series Sample:")
print(housing_median_age_series.head())
```

```python
print("\nRow using loc (index 5):")
print(df.loc[5])

print("\nRow using iloc (index 10):")
print(df.iloc[10])

print("\nSelected Columns (housing_median_age, median_income, median_house_value):")
print(df[['housing_median_age', 'median_income', 'median_house_value']].head())
```

```python
# The 'california_housing_train.csv' dataset does not contain a 'Sex' column, so renaming is not applicable.
# df.rename(columns={'Sex': 'Gender'}, inplace=True)
print("\nNo 'Sex' column to rename to 'Gender' in the current dataset.")
```

```python
print("\nHouses with housing_median_age > 30:")
print(df[df['housing_median_age'] > 30].head())

print("\nHouses with median_house_value > 300000:")
print(df[df['median_house_value'] > 300000].head())

# The original 'Female Passengers Who survived' filter is not applicable to this dataset.
# Here's an example of filtering by high median income:
print("\nHouses with median_income > 8 (indicating higher income areas):")
print(df[df['median_income'] > 8].head())
```

```python
print("\nTop 5 Highest Median House Value:")
print(df.sort_values(by='median_house_value', ascending=False).head())

print("\nAreas with Youngest Housing Median Age:")
print(df.sort_values(by='housing_median_age').head())
```

```python
print("\nAverage Median House Value by Housing Median Age:")
# Grouping by housing_median_age directly, as it has some discrete values
print(df.groupby('housing_median_age')['median_house_value'].mean().head())

print("\nAverage Population by Housing Median Age:")
print(df.groupby('housing_median_age')['population'].mean().head())

print("\nTotal Rooms by Housing Median Age (sum):")
print(df.groupby('housing_median_age')['total_rooms'].sum().head())
```
